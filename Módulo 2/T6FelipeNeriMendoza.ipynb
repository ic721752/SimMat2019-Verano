```python
import pandas as pd
import numpy as np
```

```python
i = np.arange(1,5)

df = pd.DataFrame(index=i,columns=['ai', 'bi', 'ci', 'di', 'ei', 'fi', 'pi', 'Porción generada'], dtype='float')
df.index.name = "$i$"

df['ai'] = [0.0, 0.85, 0.09, -0.09]
df['bi'] = [0.0, 0.02, -0.28, 0.28]
df['ci'] = [0.0, -0.02, 0.3, 0.3]
df['di'] = [0.25, 0.83, 0.11, 0.09]
df['ei'] = [0.0, 0.0, 0.0, 0.0]
df['fi'] = [-0.14, 1.0, 0.6, 0.7]
df['pi'] = [0.02, 0.84, 0.07, 0.07]
df['Porción generada'] = ['Tallo', 'Follaje cada vez más pequeño', 'Ramas izquierda', 'Ramas derecha']

df.round(3)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

import random
```

```python
Mat_Culcita=df.iloc[:,:-1].values
Mat_Culcita
```

```python
# Punto inicial
x=[0]
y=[0]

# Valores de i con su probabilidad
secuencia=np.array([0,1,2,3])
probabilidad=Mat_Culcita[:,-1]
a=Mat_Culcita[:,0]
b=Mat_Culcita[:,1] 
c=Mat_Culcita[:,2]
d=Mat_Culcita[:,3]
e=Mat_Culcita[:,4]
f=Mat_Culcita[:,5]

for  k in range (100000):
    i=random.choices(secuencia,probabilidad)
    x.append(a[i]*x[-1]+b[i]*y[-1]+e[i])
    y.append(c[i]*x[-2]+d[i]*y[-1]+f[i])


# Graficamos
plt.figure(figsize=(10,12))
plt.scatter(x,y,c="g",s=0.3)
plt.show()
```
